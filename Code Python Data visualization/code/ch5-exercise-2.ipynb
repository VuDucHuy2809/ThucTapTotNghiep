```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
src_file = Path.cwd() / 'data' / 'raw' / 'EPA_fuel_economy_summary.csv'
df = pd.read_csv(src_file)
```

```python
df.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.displot(data=df, x='fuelCost08', hue='year', aspect=2.5)
```

```python
sns.displot(data=df, x='fuelCost08', hue='date_range', aspect=2.5)
```

```python
sns.displot(data=df, x='fuelCost08', col='date_range')
```

```python
sns.displot(data=df, x='fuelCost08', row='date_range', kind='kde')
```

```python
sns.relplot(data=df, x='cylinders', y='comb08', kind='scatter')
```

```python
sns.relplot(data=df, x='cylinders', y='comb08', hue='drive', kind='scatter')
```

```python
sns.relplot(data=df, 
            x='cylinders', 
            y='comb08', 
            col='date_range',
            hue='drive', 
            kind='scatter')
```

```python
sns.relplot(data=df, 
            x='cylinders', 
            y='comb08', 
            col='date_range',
            hue='drive', 
            kind='line')
```

```python
sns.catplot(data=df, 
            x='date_range', 
            y='comb08', 
            kind='box', 
            showfliers=False)
```

```python
sns.catplot(data=df, 
            x='date_range', 
            y='comb08', 
            kind='boxen')
```

```python
sns.catplot(data=df, 
            x='date_range', 
            y='comb08', 
            col='fuel_type_summary',
            kind='box')
```

```python
sns.catplot(data=df, 
            x='date_range', 
            y='comb08', 
            col='fuel_type_summary',
            col_wrap=2,
            kind='box')
```

```python
sns.catplot(data=df, x='year', aspect=2,
            y='highway08', kind='point')
```

```python
sns.catplot(data=df, x='year', aspect=2,
            hue='fuel_type_summary', join=False,
            y='highway08', kind='point')
```

```python
sns.catplot(data=df, x='year', aspect=2,
            row='date_range',
            hue='fuel_type_summary',
            sharex=False,
            y='highway08', kind='bar')
```

```python
sns.catplot(data=df, 
            y='comb08', 
            x='cylinders',
            col ='drive',
            row='date_range',
            hue='transmission',
            kind='bar')
```
